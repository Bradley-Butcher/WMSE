```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
results_1 = pd.read_csv("../results/l1_results_precrecall.csv", index_col=0)
results_2 = pd.read_csv("../results/l2_results_precrecall_imb.csv", index_col=0)
```

```python
L1 = results_1[["Seed", "Samples", "L1 RECALL", "L1 V RECALL", "L1 PRECISION", "L1 V PRECISION"]]#
L1 = L1.rename(columns={"L1 RECALL": "RECALL", "L1 V RECALL": "V RECALL", "L1 PRECISION": "PRECISION", "L1 V PRECISION": "V PRECISION"})
L1["Score"] = "L1"

L2 = results_2[["Seed", "Samples", "L2 RECALL", "L2 V RECALL", "L2 PRECISION", "L2 V PRECISION"]]#
L2 = L2.rename(columns={"L2 RECALL": "RECALL", "L2 V RECALL": "V RECALL", "L2 PRECISION": "PRECISION", "L2 V PRECISION": "V PRECISION"})
L2["Score"] = "L2"

BIC = results_2[["Seed", "Samples", "BIC RECALL", "BIC V RECALL", "BIC PRECISION", "BIC V PRECISION"]]#
BIC = BIC.rename(columns={"BIC RECALL": "RECALL", "BIC V RECALL": "V RECALL", "BIC PRECISION": "PRECISION", "BIC V PRECISION": "V PRECISION"})
BIC["Score"] = "BIC"

results = BIC.append([L2])
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex = True)
fig.suptitle('Low Sample Performance on (Uniform) 10-Node Networks')
fig.set_size_inches(16.5, 10.5)
sns.set_theme(style="whitegrid")
sns.lineplot(
    data=results,
    x="Samples", y="RECALL", hue="Score", style="Score",
    markers=True, dashes=False,markersize=10,ax=ax3
)

sns.set_theme(style="whitegrid")
sns.lineplot(
    data=results,
    x="Samples", y="V RECALL", hue="Score", style="Score",
    markers=True, dashes=False,markersize=10,ax=ax4
)
sns.lineplot(
    data=results,
    x="Samples", y="PRECISION", hue="Score", style="Score",
    markers=True, dashes=False,markersize=10,ax=ax2
)

sns.set_theme(style="whitegrid")
sns.lineplot(
    data=results,
    x="Samples", y="V PRECISION", hue="Score", style="Score",
    markers=True, dashes=False,markersize=10,ax=ax1
)

ax1.set_ylabel("Skeleton Precision")
ax2.set_ylabel("V-Structure Precision")
ax3.set_ylabel("Skeleton Recall")
ax4.set_ylabel("V-Structure Recall")

ax1.set_xlim([0, 6400])
ax2.set_xlim([0, 6400])

ax3.set_xlim([0, 6400])
ax4.set_xlim([0, 6400])

ax1.set_ylim([0, 1])
ax2.set_ylim([0, 1])
ax3.set_ylim([0, 1])
ax4.set_ylim([0, 1])
```
