```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
results_1 = pd.read_csv("../results/l1_results_recall.csv", index_col=0)
results_2 = pd.read_csv("../results/l2_results_recall.csv", index_col=0)
```

```python
L1 = results_1[["Seed", "Samples", "L1 SKEL RECALL", "L1 V RECALL"]]
L1 = L1.rename(columns={"L1 SKEL RECALL": "SKEL RECALL", "L1 V RECALL": "V RECALL"})
L1["Score"] = "L1"

L2 = results_2[["Seed", "Samples", "L2 SKEL RECALL", "L2 V RECALL"]]#
L2 = L2.rename(columns={"L2 SKEL RECALL": "SKEL RECALL", "L2 V RECALL": "V RECALL"})
L2["Score"] = "L2"

BIC = results_2[["Seed", "Samples", "BIC SKEL RECALL", "BIC V RECALL"]]#
BIC = BIC.rename(columns={"BIC SKEL RECALL": "SKEL RECALL", "BIC V RECALL": "V RECALL"})
BIC["Score"] = "BIC"

results = BIC.append([L2])
```

```python
results
```

```python
sns.set_theme(style="whitegrid")
sns.lineplot(
    data=results,
    x="Samples", y="SKEL RECALL", hue="Score", style="Score",
    markers=True, dashes=False,markersize=10
)
```

```python
sns.set_theme(style="whitegrid")
sns.lineplot(
    data=results,
    x="Samples", y="V RECALL", hue="Score", style="Score",
    markers=True, dashes=False,markersize=10
)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Low Sample Performance on (Uniform) 10-Node Networks')
fig.set_size_inches(16.5, 6.5)
sns.set_theme(style="whitegrid")
sns.lineplot(
    data=results,
    x="Samples", y="SKEL RECALL", hue="Score", style="Score",
    markers=True, dashes=False,markersize=10,ax=ax1
)

sns.set_theme(style="whitegrid")
sns.lineplot(
    data=results,
    x="Samples", y="V RECALL", hue="Score", style="Score",
    markers=True, dashes=False,markersize=10,ax=ax2
)

ax1.set_ylabel("Skeleton Recall")
ax2.set_ylabel("V-Structure Recall")

ax1.set_xlim([0, 3200])
ax2.set_xlim([0, 3200])

ax1.set_ylim([0, 1])
ax2.set_ylim([0, 1])
```
